# Handling categorical variables
1. One hot encoding-creates new features and replaces the categories with either 0 or 1
2. Label encoding
3. Ordinal encoding
4. Binary encoding
5. Frequency or count encoding
6. Base-n encoder

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {'Temperature':['Hot','Cold','Very Hot','Warm','Hot','Warm','Warm','Hot','Hot','Cold'],
        'Color':['Red','Yellow','Blue','Blue','Red','Yellow','Red','Yellow','Yellow','Yellow'],
        'Target':[1,1,1,0,1,0,1,0,1,1]}
df = pd.DataFrame(data)
df1= df.copy()
df1
```

# 1. Get dummies

```python
df1= pd.get_dummies(data=df1, columns=['Temperature'],prefix='temp')
df1
```

```python
df1.info()
```

```python
b = df1.select_dtypes(include='bool').columns
df1[b]= df1[b].astype(int)
df1
```

# Instance of using label encoder
Used on categories with ordinal order- iko na mpangilio
Used for tree base models

# Instance of using one hot encoder
Its perfect for nominal categorical data where there is no inherent orders among categories
No natural order in ranking
Based on linear models,logistic progression an neural networks

# Get dummies
When using pandas and want a quick solution for using nominal variables- they have no numeric meaning and no natural ranking or order
It is usefull for regression models

# Category encoders
Useful for advanced encoding when;  1 Have many unique values and 2. want to try target encoding, hashing and binary encoding
Target encoding is when the category correlates with the target variable- risks leakage
Hashing encoding is when you have many categories and you want fewer columns
Binary encoding is for many categories

Avoid label encoder when you have nominal categories
Avoid one hot encoder and get dummies when you have too many categories
